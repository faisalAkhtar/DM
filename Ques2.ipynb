## Wine dataset

```python
import pandas as pd

file=pd.read_csv('wine.csv')
file.head()
```

```python
data=file.values
print('wine =>\n  mean  :  ',data[:,0].mean(),'\n   std  :  ',data[:,0].std())
print('Alcohol =>\n  mean  :  ',data[:,1].mean(),'\n   std  :  ',data[:,1].std())
print('malic acid =>\n  mean  :  ',data[:,2].mean(),'\n   std  :  ',data[:,2].std())
print('ash =>\n  mean  :  ',data[:,3].mean(),'\n   std  :  ',data[:,3].std())
print('acl =>\n  mean  :  ',data[:,4].mean(),'\n   std  :  ',data[:,4].std())
print('mg =>\n  mean  :  ',data[:,5].mean(),'\n   std  :  ',data[:,5].std())
print('phenol =>\n  mean  :  ',data[:,6].mean(),'\n   std  :  ',data[:,6].std())
print('flavanoid =>\n  mean  :  ',data[:,7].mean(),'\n   std  :  ',data[:,7].std())
print('nonfla. phenol =>\n  mean  :  ',data[:,8].mean(),'\n   std  :  ',data[:,8].std())
print('proanth =>\n  mean  :  ',data[:,9].mean(),'\n   std  :  ',data[:,9].std())
print('color.int =>\n  mean  :  ',data[:,10].mean(),'\n   std  :  ',data[:,10].std())
print('hue =>\n  mean  :  ',data[:,11].mean(),'\n   std  :  ',data[:,11].std())
print('OD =>\n  mean  :  ',data[:,12].mean(),'\n   std  :  ',data[:,12].std())
print('proline =>\n  mean  :  ',data[:,-1].mean(),'\n   std  :  ',data[:,-1].std())
```

```python
a=data[:,0]
b=data[:,1]
c=data[:,2]
d=data[:,3]
e=data[:,4]
f=data[:,5]
g=data[:,6]
h=data[:,7]
i=data[:,8]
j=data[:,9]
k=data[:,10]
l=data[:,11]
m=data[:,12]
n=data[:,-1]
```

```python
from sklearn import preprocessing

scaled_a=preprocessing.scale(a)
scaled_b=preprocessing.scale(b)
scaled_c=preprocessing.scale(c)
scaled_d=preprocessing.scale(d)
scaled_e=preprocessing.scale(e)
scaled_f=preprocessing.scale(f)
scaled_g=preprocessing.scale(g)
scaled_h=preprocessing.scale(h)
scaled_i=preprocessing.scale(i)
scaled_j=preprocessing.scale(j)
scaled_k=preprocessing.scale(k)
scaled_l=preprocessing.scale(l)
scaled_m=preprocessing.scale(m)
scaled_n=preprocessing.scale(n)
```

```python
print('_____Standardized values_____\n')
print('wine =>\n  mean  :  ',int(scaled_a.mean()),'\n   std  :  ',scaled_a.std())
print('Alcohol =>\n  mean  :  ',int(scaled_b.mean()),'\n   std  :  ',scaled_b.std())
print('malic acid =>\n  mean  :  ',int(scaled_c.mean()),'\n   std  :  ',scaled_c.std())
print('ash =>\n  mean  :  ',int(scaled_d.mean()),'\n   std  :  ',scaled_d.std())
print('acl =>\n  mean  :  ',int(scaled_e.mean()),'\n   std  :  ',scaled_e.std())
print('mg =>\n  mean  :  ',int(scaled_f.mean()),'\n   std  :  ',scaled_f.std())
print('phenol =>\n  mean  :  ',int(scaled_g.mean()),'\n   std  :  ',scaled_g.std())
print('flavanoid =>\n  mean  :  ',int(scaled_h.mean()),'\n   std  :  ',scaled_h.std())
print('nonfla. phenol =>\n  mean  :  ',int(scaled_i.mean()),'\n   std  :  ',scaled_i.std())
print('proanth =>\n  mean  :  ',int(scaled_j.mean()),'\n   std  :  ',scaled_j.std())
print('color.int =>\n  mean  :  ',int(scaled_k.mean()),'\n   std  :  ',scaled_k.std())
print('hue =>\n  mean  :  ',int(scaled_l.mean()),'\n   std  :  ',scaled_l.std())
print('OD =>\n  mean  :  ',int(scaled_m.mean()),'\n   std  :  ',scaled_m.std())
print('proline =>\n  mean  :  ',int(scaled_n.mean()),'\n   std  :  ',scaled_n.std())
```
